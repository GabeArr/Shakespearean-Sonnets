```python
import gensim
from gensim import corpora, models, similarities
```

```python
##load this pretrained model in Gensim and you should be good to go
model = gensim.models.KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin', binary=True)
```

#### Q1: Find the cosine similarities between the following word pairs/tuples: 
* (France, England)
* (smaller, bigger)
* (england, london)
* (france, rocket)
* (big, bigger)


```python
france_england = model.similarity('France', 'England')
smaller_bigger = model.similarity('smaller', 'bigger')
england_london = model.similarity('england', 'london')
france_rocket = model.similarity('france', 'rocket')
big_bigger = model.similarity('big', 'bigger')
```

```python
import pandas as pd
d = {'word pairs':[france_england,smaller_bigger,england_london,
                   france_rocket, big_bigger]}
pd.DataFrame(d)
```

#### Q2 : Extract the vector representations of the words:
* France
* England
* smaller
* bigger
* London
* big

```python
vector_list = []
for word in ['France', 'England', 'smaller', 'bigger', 'London', 'rocket', 'big']:
    vector_list.append((word, model[word]))
print vector_list
```

#### Q3:  Now using Numpy extract the magnitude (L2 norm) of the vectors and subtract among the word pairs in Q1. Do the magnitudes of the vectors matter? Justify your answer?

```python
import numpy as np
magnitudes = {}
for vector in vector_list:
    x = np.array(vector[1])
    magnitudes[vector[0]] = np.linalg.norm(x)
```

```python
france_subtract_england = magnitudes['France'] - magnitudes['England']
smaller_subtract_bigger = magnitudes['smaller'] - magnitudes['bigger']
england_subtract_london = magnitudes['England'] - magnitudes['London']
france_subtract_rocket = magnitudes['France'] - magnitudes['rocket']
big_subtract_bigger = magnitudes['big'] - magnitudes['bigger']
d = {'subtracted word pairs':[france_subtract_england,smaller_subtract_bigger,england_subtract_london,
                   france_subtract_rocket, big_subtract_bigger]}
pd.DataFrame(d)
```

```python
import nltk
import string

from urllib2 import urlopen
# from urllib.request import urlopen
import nltk, re, pprint
from nltk import word_tokenize
# sent_tokenizer=nltk.data.load('tokenizers/punkt/english.pickle')
from nltk import tokenize
```

```python
### Download and load  "The Importance of Being Earnest A Trivial Comedy for Serious People" by Oscar 
## Wilde from Project Gutenberg : https://www.gutenberg.org

nltk.download('punkt')

url = "http://www.gutenberg.org/cache/epub/1041/pg1041.txt" ## Your raw text file location 
resp = urlopen(url)
raw = resp.read().decode('utf8')
firstlook = tokenize.sent_tokenize(raw)
```

```python
pattern = r'''(?x)  # set flag to allow verbose regexps
(?:[A-Z]\.)+        # abbreviations, e.g. U.S.A.
|\w+(?:[-']\w+)*    # words with optional internal hyphens
|\$?\d+(?:\.\d+)?   # currency, e.g. $12.80 
|\.\.\.             # elipses
|[.,;"'?()-_`]      # these are separate tokens
'''
#print(nltk.regexp_tokenize(raw,pattern))
tokenized_raw =" ".join( nltk.regexp_tokenize(raw,pattern))
tokenized_raw=tokenize.sent_tokenize(tokenized_raw)
```

```python
len(tokenized_raw[:10])
```

```python
nopunct=[]
for sent in tokenized_raw:
    a=[w for w in sent.split() if w not in string.punctuation]
    nopunct.append(" ".join(a))
#create a set of stopwords
tok_corp= [nltk.word_tokenize(sent) for sent in nopunct]
```

```python
### creating a list of unique words 

combined_list=[" ".join(w) for w in tok_corp]
unique_list=[]
for sent in combined_list:
    unique_list.append([w for w in sent.split()])
unique_list=list(set([item for sublist in unique_list for item in sublist]))
```

```python
unique_words=unique_list
```

```python
### Its just one single command
model = gensim.models.Word2Vec(tok_corp, min_count=1, size = 16, window=5)

## Window size : The parameter that fixes the length of window of words by which the model sweeps 
# the data

## Size: Size of the vector

## min_count: accept a word if it has a certain minimum number of occurances in the document
```

```python
## Extracting the respective vectors corresponding to the words

vector_list=[] ## n by d matrix contaiing words and their respective vectors
for word in unique_words:
    vector_list.append(model[word])
```

```python
## A popular non-linear dimensionality reduction technique that preserves greatly thge local 
## and global structure of the data. Essentially tries to reconstruct the subspace in which the 
## data exists
## We will use the Sklearn TSNE implementation 

from sklearn.manifold import TSNE

# Lets dim reduce the 16 dimension vectors to 2dimensions to vizualise the dataset 
data_embed=TSNE(n_components=2, perplexity=50, verbose=2, method='barnes_hut').fit_transform(vector_list)


## Parameters
## n_components = number of dimensions you want your data to be reduced
## preplexity =  Number of neighboours to fit the gaussian , normally 30 
```

```python
## Vizualise
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
x_axis= data_embed[:,0]
y_axis=data_embed[:,1]

plt.scatter(x_axis, y_axis)
plt.show() ## The plots vary each time you run them 
```
